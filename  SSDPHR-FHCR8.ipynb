# Question 4.
Provide two numpy-arrays, one for Italy and one for Japan, with the total number of deaths by age group from all causes

# Question 6.
 Looking specifically at the ICD10-category for Neoplasm (C00-D48). What were the top five causes of deaths caused by Neoplasm (top five ICD-10 terms) in Italy and Japan across all years (in absolute numbers). 

```python
#Concatenating WHO Mortality data into one DataFrame
ICD_dataframes = [df_ICD_1, df_ICD_2]
complete_ICD_data= pd.concat(ICD_dataframes)
complete_ICD_data.head()
```

```python
#Italy's Country code is 4180 as identified above.
complete_italy_data = complete_ICD_data[complete_ICD_data['Country'] == 4180]
complete_italy_data.tail()
```

```python
#Creating an ICD code list for all deaths caused by Neoplasms, starting with all codes starting with "C"
neoplasms_code_list = []
for code_C in range(0,98):
    if code_C <10:
        neoplasms_code_list.append("C"+"0"+str(code_C))
    else:
        neoplasms_code_list.append("C"+str(code_C)) 
neoplasms_code_list
```

```python
#Adding all of the ICD 10 codes for neoplasms that start with "D"
for code_D in range(0,49):
    if code_D <10:
        neoplasms_code_list.append("D"+"0"+str(code_D))
    else:
        neoplasms_code_list.append("D"+str(code_D))
neoplasms_code_list        
```

C89 code has not been attributed to a Neoplasm, therefore this will be removed from the list.

```python
#Removing code "C89"
neoplasms_code_list.remove("C89")
```

```python
#Changing the list of codes into a tuple
neoplasms_code_tuple = tuple(neoplasms_code_list)
```

```python
#Selecting all deaths caused by Neoplasms
italy_np_data = complete_italy_data[complete_italy_data['Cause'].str.startswith(neoplasms_code_tuple)]
italy_np_data.head()
```

```python
#Calculating the total number of the deaths for each cause 
top_italy_np = italy_np_data.groupby('Cause')['Deaths1'].sum().sort_values(ascending=False).head()
#Reading the results into a DataFrame
top_italy_np = top_italy_np.to_frame()
#Displaying the DataFrame
top_italy_np
```

```python
#Japan's Country code is 3160 as identified above.
complete_japan_data = complete_ICD_data[complete_ICD_data['Country'] == 3160]
complete_japan_data.tail()
```

```python
#Selecting all deaths caused by Neoplasms
japan_np_data = complete_japan_data[complete_japan_data['Cause'].str.startswith(neoplasms_code_tuple)]
japan_np_data.head()
```

```python
#Calculating the total number of the deaths for each cause 
top_japan_np = japan_np_data.groupby('Cause')['Deaths1'].sum().sort_values(ascending=False).head()
#Reading the results into a DataFrame
top_japan_np = top_japan_np.to_frame()
#Displaying the DataFrame
top_japan_np
```

# Question 7. 
For the top five deaths caused by Neoplasm calculate the proportion in relation to the total number of deaths.

```python
#Calculating the proportions for the Top 5 Neoplasm causes
top_italy_np['Proportion of overall deaths'] = round((top_italy_np['Deaths1']/[complete_italy_data.Deaths1.sum()])*100,2)
top_italy_np
```

```python
#Calculating the proportions for the Top 5 Neoplasm causes
top_japan_np['Proportion of overall deaths'] = round((top_japan_np['Deaths1']/[complete_japan_data.Deaths1.sum()])*100,2)
top_japan_np
```
