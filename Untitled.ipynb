```python
#Reading the csv file with the Country codes into a DataFrame
Country_codes = pd.read_csv('2 country_codes.csv')
Country_codes.head()
```

```python
#Reading the csv file with information about the Population of the countries into a DataFrame
Population  = pd.read_csv('1 pop.csv')
Population.head()
```

```python
#Creating a new DataFrame with data only from the year 2010
Population_2010 = Population[Population['Year'] == 2010]
Population_2010.head()
```

```python
#Identifying the country codes for Italy, Iceland and New Zealand and finding out the population of these countries in 2010.
list_of_countries = ['Italy', 'Japan']
list_of_codes = []
for country in list_of_countries:
    Country_code = Country_codes.loc[Country_codes['name'] == country, 'country'].iloc[0]
    Country_code 
    print('Country code for ' + country + ' is : ' + str(Country_code))
    list_of_codes.append(Country_code)
    Country_pop = Population_2010.loc[Population_2010['Country'] == Country_code, 'Pop1'].iloc[0]
    print('Population of ' + country + ' in 2010 was ' + str(Country_pop))
```

```python
#Selecting all deaths caused by Neoplasms
Italy_Neoplasm_data = Complete_Italy_data[Complete_Italy_data['Cause'].str.startswith(Neoplasms)]
Italy_Neoplasm_data.head()
```

```python
#Calculating the total number of the deaths for each cause 
Top_5_Italy_Neoplasm = Italy_Neoplasm_data.groupby('Cause')['Deaths1'].sum().sort_values(ascending=False).head()
#Reading the results into a DataFrame
Top_5_Italy_Neoplasm = Top_5_Italy_Neoplasm.to_frame()
#Displaying the DataFrame
Top_5_Italy_Neoplasm
```

```python
#Using the .sum() method to calculate the total number of deaths in Italy across all the years
Complete_Italy_data.Deaths1.sum()
```

```python
#Calculating the proportion of overall deaths (as a percentage)
Top_5_Italy_Neoplasm['Proportion of overall deaths'] = (Top_5_Italy_Neoplasm['Deaths1']/13012870)*100
Top_5_Italy_Neoplasm
```

```python
#Rounding the Proportion of overall deaths to 2 decimal places places
for Proportion in Top_5_Italy_Neoplasm['Proportion of overall deaths']:
    print(round(Proportion,2)) 
```
