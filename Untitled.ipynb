Find the country code for Italy and Japan.

```python
#Identifying the country codes for Italy and Japan
list_of_countries = ['Italy', 'Japan']
list_of_codes = []
for country in list_of_countries:
    code_of_country = country_codes.loc[country_codes['name'] == country, 'country'].iloc[0] 
    list_of_codes.append(code_of_country)
    print('Country code for ' + country + ' is: ' + str(code_of_country))
       
```

# Question 2.
In the year 2010 how many people died (from all causes) in Italy and Japan

```python
#Selecting records for the year 2010
ICD_2010 = df_ICD_2[df_ICD_2['Year'] == 2010]
ICD_2010.head()
```

```python
#Determining number of deaths in Italy and Japan in 2010
dict_of_countries = {'4180':'Italy', '3160':'Japan'}
for key, value in dict_of_countries.items():
    death_data = ICD_2010[ICD_2010['Country'] == int(key)]
    total_number_of_deaths = death_data.Deaths1.sum()
    print('Total number of deaths from all causes in ' + value + ' in 2010 was ' + str(total_number_of_deaths))
```

```python
age_groups = italy_age_group_df['age_groups']
number_of_deaths = italy_age_group_df['number_of_deaths']

plt.style.use('classic')
plt.rcParams['figure.figsize'] = [10, 5]
plt.bar(age_groups, number_of_deaths, color = "olivedrab", edgecolor = 'gold', width = 0.6)
plt.title("Number of deaths in Italy per age group")
plt.xlabel("Age groups")
plt.ylabel("Number of deaths")
# Add a legend & grid
plt.legend("Number of deaths", loc=2, prop={'size':12})
plt.grid(True)
plt.xticks(ticks=np.arange(10), labels=age_groups, rotation=40)

plt.show()
```
